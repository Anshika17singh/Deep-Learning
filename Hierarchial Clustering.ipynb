```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
# Temperature data
x=np.array([[35,20],[40,23],[32,19],[50,30],[52,33],[10,3],[24,12],[8,0],[7,-2],[3,0]])
```

```python
x=x.reshape[-1,2]
```

```python
plt.scatter(x[:,0],x[:,1],color='blue')
plt.title('Temperature Report')
plt.xlabel('High Temperature')
plt.ylabel('low Temperature')
plt.show()
```

```python
abc= AgglomerativeClustering(n_clusters=3)
```

```python
prediction=abc.fit_predict(x)
```

```python
prediction
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=abc,palette='Oranges')
plt.title('Temperature Report')
plt.xlabel('High Temperature')
plt.ylabel('low Temperature')
plt.show()
```

```python
li=linkage(x,method='ward')
plt.figure(figsize=(5,4))
dendrogram(li,orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.title('Temperature')
plt.show()
```
